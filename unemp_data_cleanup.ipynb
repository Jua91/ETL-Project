```python
import pandas as pd
```

```python
url = "https://www.bls.gov/lau/lastch18.htm"
tables = pd.read_html(url)
```

```python
tables[0]
```

```python
unemp_df = tables[0]
unemp_df.columns = ["State","2017_unemployment_rate","2018_unemployment_rate","2017-2018_unemp_change","rank"]
unemp_df = unemp_df.drop([1,53,54])
unemp_df = unemp_df.drop(columns=['rank'])
unemp_df = unemp_df.reset_index(drop =True)
unemp_df
```

```python
unemp_df.to_csv("../Data/unemployment(2017-2018).csv", index=False)
```
