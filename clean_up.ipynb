```python
import pandas as pd
from sqlalchemy import create_engine
```

# Store CSV into DataFrame

```python
csv_file1 = "Resources/gdp_dollars(2017-2018).csv"
gdpd_df = pd.read_csv(csv_file1)
gdpd_df.head()
```

```python
# gdp1_df.drop(gdp1_df.index[0])
# gdp1_df.head()
gdp1_df = gdpd_df.drop(0)
gdp1_df.head()
```

```python
csv_file2 = "Resources/gdp_growth(2017-2018).csv"
gdpg_df = pd.read_csv(csv_file2)
gdpg_df.head()
```

```python
gdp2_df = gdpg_df.drop(0)
gdp2_df.head()
```

```python
merge_gdp = pd.merge(gdp1_df, gdp2_df, on="GeoName")
merge_gdp.head()
```

```python
merge_gdp = merge_gdp.rename(columns={"GeoName": "State","2017":"2017_gdp_rate","2018":"2018_gdp_rate","2017-2018":"2017-2018_gdp_change"})
merge_gdp.head()
```

# Create new data with select columns

```python
gdp_data_df = merge_gdp[['State', '2017_gdp_rate', '2018_gdp_rate', '2017-2018_gdp_change']].copy()
gdp_data_df.head()
```

```python
gdp_data_df.to_csv("Resources/GDP_Final(2017-2018).csv", index=False)
```

```python
url = "https://www.bls.gov/lau/lastch18.htm"
```

```python
tables = pd.read_html(url)
tables
```

```python
unemp_df = tables[0]
unemp_df.head()
```

```python
unemp_df.columns = ["State","2017_unemployment_rate","2018_unemployment_rate","2017-2018_unemp_change","rank"]
unemp_df = unemp_df.drop([1,53,54])
unemp_df = unemp_df.drop(columns=['rank'])
unemp_df = unemp_df.reset_index(drop =True)
unemp_df
unemp_data_df = unemp_df.drop(0)
unemp_data_df.head()
```

```python
# unemp_data_df.sort_values(by='State',ascending=False,inplace=False)
unemp_final_df= unemp_data_df.sort_values(by='State',inplace=False)
unemp_final_df.head()
```

```python
unemp_final_df.to_csv("Resources/unemployment(2017-2018).csv", index=False)
```
