```python
import pandas as pd
```

# Unemployment Data

```python
url = "https://www.bls.gov/lau/lastch18.htm"
tables = pd.read_html(url)
```

```python
tables[0].head()
```

```python
unemp_df = tables[0]
unemp_df.columns = ["State","2017_unemployment_rate","2018_unemployment_rate","2017-2018_unemp_change","rank"]
unemp_df = unemp_df.drop([0,1,53,54])
unemp_df = unemp_df.drop(columns=['rank'])
unemp_df = unemp_df.reset_index(drop =True)
unemp_final_df = unemp_df.sort_values("State").reset_index(drop=True)
unemp_final_df.head()
```

# GDP Data

```python
gdp_file1 = "../Data/gdp_dollars(2017-2018).csv"
gdpd_df = pd.read_csv(gdp_file1,header=4)
gdpd_df.head()
```

```python
gdp1_df = gdpd_df.drop(0)
gdp1_df.head()
```

```python
gdp_file2 = "../Data/gdp_growth(2017-2018).csv"
gdpg_df = pd.read_csv(gdp_file2,header=4)
gdpg_df.head()
```

```python
gdp2_df = gdpg_df.drop(0)
gdp2_df.head()
merge_gdp = pd.merge(gdp1_df, gdp2_df, on="GeoName")
merge_gdp.head()
```

```python
merge_gdp = merge_gdp.rename(columns={"GeoName": "State","2017":"2017_gdp_dollars","2018":"2018_gdp_dollars","2017-2018":"2017-2018_gdp_change"})
merge_gdp.head()
gdp_data_df = merge_gdp[['State', '2017_gdp_dollars', '2018_gdp_dollars', '2017-2018_gdp_change']].copy()
gdp_data_df.head()
```

```python
#gdp_data_df.to_csv("../Data/GDP_Final(2017-2018).csv", index=False)
```

# Health Insurance Data

```python
hi_file = "../Data/health_insurance(2017-2018).csv"
hi_df = pd.read_csv(hi_file)
hi_df.head()
```

```python
# Separate the tables for each year
hi_2018_df = hi_df.drop_duplicates("Name", keep="first")
hi_2018_df = hi_2018_df.reset_index(drop=True)

hi_2017_df = hi_df.drop_duplicates("Name", keep="last")
hi_2017_df = hi_2017_df.reset_index(drop=True)

hi_2018_reduced_df = hi_2018_df[["Name","Uninsured: %"]]
hi_2017_reduced_df = hi_2017_df[["Name","Uninsured: %"]]
hi_combined_df = pd.merge(hi_2017_reduced_df,hi_2018_reduced_df, on="Name", suffixes=(' 2017',' 2018'))
hi_combined_df.head()
```

```python
hi_final_df = hi_combined_df.rename(columns={"Name":"State","Uninsured: % 2017":"2017_uninsured(%)","Uninsured: % 2018":"2018_uninsured(%)"})
hi_final_df.head()
```

```python
hi_rate_change = []
for index, row in hi_final_df.iterrows():
    hi_rate_change.append(row["2018_uninsured(%)"] - row["2017_uninsured(%)"])

hi_final_df["2017-2018_uninsured_rate_change"] = hi_rate_change
hi_final_df.head()
```

# Merge all three data

```python
gdp_unemp_df = pd.merge(gdp_data_df,unemp_df, on="State")
final_combined_df = pd.merge(gdp_unemp_df,hi_final_df, on="State")
final_combined_df.head()
```

```python
final_combined_df.to_csv("../Data/Final_data.csv", index=False)
```
